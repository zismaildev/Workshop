# Generate Embeddings for Mobile Reviews using Ollama

This notebook reads `mobile_reviews.json`, generates vector embeddings for the `review_text` using a local Ollama model, and saves the result.

```python
# Install the Ollama Python library if not already installed
%pip install ollama
```

```python
import json
import ollama
import os

# Configuration
INPUT_FILE = 'mobile_reviews.json'
OUTPUT_FILE = 'mobile_reviews_with_embeddings.json'
EMBEDDING_MODEL = 'qwen3-embedding'  # Ensure you have pulled this model: `ollama pull qwen3-embedding`
```

```python
# Load the dataset
try:
    with open(INPUT_FILE, 'r') as f:
        reviews = json.load(f)
    print(f"Loaded {len(reviews)} reviews from {INPUT_FILE}")
except FileNotFoundError:
    print(f"Error: {INPUT_FILE} not found. Make sure you are in the correct directory.")
```

```python
# Generate embeddings
print(f"Generating embeddings using model: {EMBEDDING_MODEL}...")

for i, review in enumerate(reviews):
    text_to_embed = review.get('review_text', '')
    
    if text_to_embed:
        try:
            response = ollama.embeddings(model=EMBEDDING_MODEL, prompt=text_to_embed)
            review['review_embedding'] = response['embedding']
            print(f"[ {i+1}/{len(reviews)} ] Embedded review {review.get('review_id', 'unknown')}")
        except Exception as e:
            print(f"Error embedding review {i}: {e}")
    else:
        print(f"Skipping review {i} (no text found)")

print("Embedding generation complete.")
```

```python
# Preview one record to verify structure
if reviews:
    print(json.dumps(reviews[0], indent=2))
```

```python
# Save the augmented dataset
with open(OUTPUT_FILE, 'w') as f:
    json.dump(reviews, f, indent=2)
    
print(f"Saved data with embeddings to {OUTPUT_FILE}")
```
